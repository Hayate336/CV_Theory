```python
font = cv2.FONT_HERSHEY_SIMPLEX
font_size = 1.1
font_thickness = 4
x,y = 350,50
```

```python
image = cv2.imread('/content/color.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
comp = cv2.bitwise_not(image)
(thresh, bw) = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

image = cv2.putText(image, 'Original Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
comp = cv2.putText(comp, 'Complemented Image', (x,y), font, font_size, (255,255,255), font_thickness, cv2.LINE_AA)
gray = cv2.putText(gray, 'Grayscale Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
bw = cv2.putText(bw, 'Binary Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)

H1 = np.concatenate((image, comp), axis=1)
H2 = np.concatenate((gray, bw), axis=1)
cv2_imshow(H1)
cv2_imshow(H2)
```
